```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
import  plotly.plotly  as  py 
import seaborn as sns
```

```python
pob_eeuu = pd.read_excel('poblacionfEEUU.xlsx')
pob_eeuu.head(10)
```

```python
pob_eeuu.tail(15)
```

```python
#para eliminar los NAN
poblacion=pob_eeuu.dropna(how='all')
print(poblacion)
```

```python
print('Table 1. Annual Estimates of the Resident Population for the United States, Regions, States, and Puerto Rico: April 1, 2010 to July 1, 2016 \n%s' % poblacion)
```

```python
pobfinal=poblacion.iloc[:,1:].values

#Para comparar con el PBI voy a trabajar con la poblacion desde el 2011
#pobval=pob_eeuu.iloc[:, 2:].values
print(pobfinal)
pobfinal.shape
```

```python
area=poblacion.iloc[:, :1].values

print(area)
len(area)
```

```python
poblacion.info()

print('  AÑO    MEDIA     DESVIACION ')
pob_meana= np.around(pobfinal.mean(axis=0), decimals=2)
pob_sda= np.around(pobfinal.std(axis=0), decimals=2)
list(zip(poblacion.columns[1:],  pob_meana,    pob_sda))
```

```python
print('  CIUDAD       MEDIA     DESVIACION ')
pob_meancit= np.around(pobfinal.mean(axis=1), decimals=2)
pob_sdcit= np.around(pobfinal.std(axis=1), decimals=2)
list(zip(area ,pob_meancit,pob_sdcit))
```

```python
# Pasamos el dataset desde un formato "ancho" a un formato "largo"
formatted_pob = pd.melt(poblacion, ['Geographic Area'], var_name="Año",value_name="POBLACION")
```

```python
sns.set(style = "darkgrid")
sns.boxplot(x = 'Año', y = 'POBLACION', hue = 'Año', orient='v',data = formatted_pob, palette= 'Set1')
legend = []
plt.legend(legend)
plt.show()
```

```python
sns.violinplot(x = 'Año', y = 'POBLACION', hue = 'Año', orient='v',data = formatted_pob, palette= 'Set2')
legend = []
plt.legend(legend)
plt.show()
#el punto medio en el grafico es la media
```
